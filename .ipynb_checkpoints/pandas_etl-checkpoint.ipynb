### Extract CSVs into DataFrames

```python
# Creates lists for data to be stored 
id = []
name = []
weight = []
height = []
bred_for = []
breed_group = []
url = r"https://api.thedogapi.com/v1/breeds"
response = requests.get(url).json()

pprint(response[0])
```

```python

for i in response:
    if i['id'] == 21: break 
    print(f"{i['id']}, breed = {i['name']}, weight(Kg) {i['weight']['metric']}, height (cm) {i['height']['metric']}")
    
```

```python
#appending the data to the empty lists 
for i in response:
    id.append(i["id"])
    name.append(i["name"])
    weight.append(i["weight"]['metric'])
    height.append(i["height"]['metric'])
    if 'breed_group' in i:
        breed_group.append(i['breed_group'])
    else:
        breed_group.append('Unknown')
    if 'bred_for' in i:
        bred_for.append(i['bred_for'])
    else:
        bred_for.append('Unknown')
        
#    bred_for = []
#breed_group = []
print(len(bred_for))
print(len(breed_group))
# create a dictionary for establishing dataframe
dog_data = {
    "ID":id,
    "Breed":name,
    "Weight": weight,
    "Height": height,
    "breed_group": breed_group,
    "bred_for":bred_for,
}

dog_df = pd.DataFrame(dog_data)

dog_df
```

```python
intelligence = "Resources/intelligence.csv"
intelligence_df = pd.read_csv(intelligence)
intelligence_df.head(30)
```

```python
# using merge function by setting how='inner'
final_df = pd.merge(intelligence_df, dog_df, 
                   on='Breed', 
                   how='inner')
final_df
```

```python
final_df2 = final_df[['Breed','Classification','Weight','Height','breed_group', 'bred_for']]
final_df2.head(30)
```
